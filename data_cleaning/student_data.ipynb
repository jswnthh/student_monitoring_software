```python
import pandas as pd
import hashlib
import barcode
from barcode.writer import ImageWriter
from barcode import Code128
import os   
import code128
```

```python
df_gen = pd.read_csv("gen.csv")
```

```python
df_gen['dept'] = 'GEN'
```

```python
df_gen.head()
```

```python
df_vc = pd.read_csv("vc.csv")
```

```python
df_vc['dept'] = 'VC'
```

```python
df_vc.head()
```

```python
df_vc.drop(['guardian_name', 'guardian_num'], axis=1, inplace=True)
```

```python
df_vc.head()
```

```python
df_vc = df_vc.rename(columns={
    'stud_num': 'stud_no',
})
```

```python
df = pd.concat([df_bca, df_gen, df_vc], ignore_index=True)
```

```python
df.sample(20)
```

```python
df.drop(['barcode_hash'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from PIL import Image

# Generate barcode images for each roll_no in df with 3cm x 1cm size


output_dir = "barcodes"
os.makedirs(output_dir, exist_ok=True)

for roll_no in df['roll_no']:
    barcode_img = code128.image(roll_no, height=100, thickness=3)  # 100px height, thickness=3 for clarity
    # Resize to 3cm x 1cm at 300dpi: 1cm = 118px, 3cm = 354px
    barcode_img = barcode_img.resize((354, 118), Image.LANCZOS)
    barcode_img.save(os.path.join(output_dir, f"{roll_no}.png"))
```
