```python
import pandas as pd
import hashlib
import barcode
from barcode.writer import ImageWriter
from barcode import Code128
import os   
import code128
```

```python
df = pd.read_csv('student_data2.csv')
```

```python
df.sample(10)
```

```python
# Create hash function
def generate_hash(roll_no):
    return hashlib.sha256(roll_no.encode()).hexdigest()[:10]  # short version (10 chars)

# Apply to DataFrame
df['barcode_hash'] = df['roll_no'].astype(str).apply(generate_hash)
```

```python
df.sample(10)
```

```python
os.makedirs('barcodes', exist_ok=True)
```

```python
for index, row in df.iterrows():
    value = str(row['barcode_hash'])   # or row['roll_no']
    filename = f"barcodes/barcode_{row['roll_no']}.png"

    # Generate and save PNG barcode
    img = code128.image(value)
    img.save(filename)
```

```python
new_order = [
    'barcode_hash',  # You want this first now?
    'roll_no',
    'student_name',
    'DOB',
    'gender',
    'student_no',
    'blood_group',
    'parent_name',
    'parent_no'
]

# Rearrange columns
df = df[new_order]

# Save or display
print(df.head())
```

```python
df.to_csv("student_data2.csv", index=False)
```
